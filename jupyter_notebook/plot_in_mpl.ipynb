```python
x=np.linspace(0,10,100)
```

```python
for_save=plt.figure()
plt.plot(x,np.sin(x),"--")
plt.plot(x,np.cos(x),"*")
```

```python
# Сохранение в файл
for_save.savefig("my_first_paint.png");
```
