```python
url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year={year}&Month={month}&timeframe=1&submit=Download+Data"
```

```python
url = url_template.format(month=3, year=2012)
weather_mar2012 = pd.read_csv(url,
                              index_col='Date/Time',
                              parse_dates=True,
                              encoding='latin1')
```

```python
weather_mar2012["Temp (Â°C)"].plot(figsize=(15, 5))
```

```python
weather_mar2012 = weather_mar2012.dropna(axis=1, how='any')
weather_mar2012
```

```python
weather_mar2012 = weather_mar2012.drop(
    ["Latitude (y)", "Station Name", "Climate ID", "Year", "Month"], axis=1)
```

```python
weather_mar2012
```

```python
weather_mar2012.columns = ("dell", "day", "time", "temp", "dew_temp",
                           "rel_hum", "wind_spd", "visibility", "stn_press",
                           "weather")
```

```python
weather_mar2012
```

```python
weather_mar2012 = weather_mar2012.drop(["dell"], axis=1)
```

```python
weather_mar2012
```

```python
tempers = weather_mar2012[["temp"]].copy()
```

```python
tempers
```

```python
tempers.loc[:, "Hour"] = weather_mar2012.index.hour
```

```python
tempers.groupby("Hour").aggregate(np.median).plot()
```
