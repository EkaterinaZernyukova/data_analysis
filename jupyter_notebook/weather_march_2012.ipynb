```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year={year}&Month={month}&timeframe=1&submit=Download+Data"
```

```python
url = url_template.format(year = 2012, month = 5)
url
```

```python
weather_mar2012 = pd.read_csv(url,index_col='Date/Time',
                              parse_dates=True,
                              encoding='latin1')
```

```python
weather_mar2012
```

```python
weather_mar2012 = weather_mar2012.dropna(axis=1, how="any")
weather_mar2012
```

```python
weather_mar2012["Station Name"].value_counts()
```

```python
weather_mar2012 = weather_mar2012.drop(
    ["Latitude (y)", "Station Name", "Climate ID", "Year", "Month"], axis=1)
weather_mar2012
```

```python
weather_mar2012.columns = [
    "dell", "day", "time", "temp", "dew_temp", "rel_num", "wind", "visibility",
    "stn_press", "weather"
]
```
