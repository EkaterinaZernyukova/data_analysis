```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = np.loadtxt("../resources/cardio_train.csv", delimiter=";", skiprows=1)
df.shape
```

## Типы визуализации для одной переменной (одного столбца данных).

```python
height = df[:100, 3]  # рост первых 100 человек
height
```

Simple plot (точки соединяются)

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.plot(height)
```

Scatter plot (точки не соединяются)

```python
plt.scatter(df[:100, 0], height, marker="*", s=80)
# обязательно два аргумента x, y!
```

Bar plot (столбчатые диаграммы): bar, barh

```python
id = df[:100, 0]
fig, ax = plt.subplots(figsize=(15, 5))
plt.bar(id, height=height)
```

```python
fig, ax = plt.subplots(figsize=(5, 15))
plt.barh(id, width=height)
```

Гистограмма

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].hist(height)
ax[1].hist(height, histtype="step")
```

```python
plt.hist(height, bins=15)
```

Box plot 

```python
fig.ax = plt.subplots(figsize=(10, 5))
plt.boxplot(height, vert=False);
```

```python
# Сравним мужчин и женщин по росту
women_height=df[df[:,2]==1,3]
men_height=df[df[:,2]==2,3]
fig,ax=plt.subplots(figsize=(15,5))
plt.boxplot([women_height,men_height],vert=False,labels=["women","men"]);
```

Violin plot 

```python
fig,ax=plt.subplots(figsize=(15,5))
plt.violinplot([women_height,men_height],vert=False);
```

## Визуализации двух переменных (столбцов) совместно.

Scatter plot

```python
height=df[:100,3]
age=df[:100,1]/365.25 # изначально возраст был в днях

fig,ax=plt.subplots(figsize=(15,5))
plt.scatter(age,height)
plt.xlabel("age,days")
plt.ylabel("height,cm")
plt.title("Age vs. Height");
```

Bar plot

```python
fig,ax=plt.subplots()
plt.bar([1,2],height=[women_height.mean(), men_height.mean()],color=["red","blue"])
ax.set_xticks([1,2])
ax.set_xticklabels(["women","men"]);
```
