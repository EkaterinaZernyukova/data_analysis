```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Подписи осей: set_xlabel(), set_ylabel()

Название графика: set_title()

```python
grafik, ax = plt.subplots()
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, "-.")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title(r"График $y=\sin x$")
grafik.savefig("sin.png", dpi=300)
# dpi = dots per inch (разрешение картинки)
```

```python
img = plt.imread("sin.png")  # считать рисунок
plt.imshow(img)
# отобразить рисунок
```

Цвет: color=...

```python
plt.plot(x, np.sin(x - 0), color="c")
# short color code (rgbcmykw)
```

```python
plt.plot(x, np.sin(x - 0), color="0.80")
# Grayscale between 0 and 1
```

```python
plt.plot(x, np.sin(x - 0), color="#11FF88")
# Hex code (RRGGBB from 00 to FF)
```

```python
plt.plot(x, np.sin(x - 0), color=(0.5, 0.2, 0.9))
# RGB tuple, values 0 to 1
```

```python
plt.plot(x, np.sin(x - 6), color="chartreuse")
# all HTML color names supported
```

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='c')
plt.plot(x, np.sin(x - 2), color='0.75')
plt.plot(x, np.sin(x - 3), color='#FFDD44')
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

Стиль линии: linestyle=...

```python
plt.plot(x, x + 0, linestyle="solid")
plt.plot(x, x + 1, linestyle="dashed")
plt.plot(x, x + 2, linestyle="dashdot")
plt.plot(x, x + 3, linestyle="dotted")
```

Толщина линии: linewidth=...

```python
for iw in range(1, 12):
    plt.plot(x, x + iw, linewidth=iw)
```

Тип маркера: marker=...

```python
x = np.arange(1, 11)
plt.plot(x, np.sin(x), marker="$K$")
plt.plot(x, np.cos(x), marker=">")
```

Размер маркера: markersize=...

```python
x = np.arange(1, 11)
plt.plot(x, np.sin(x), marker="*", markersize=10, label="sin(x)")
plt.plot(x, np.sin(x + 1), marker="H", markersize=10, label="sin(x+1)")
# plt.axis("equal")
plt.legend(loc=4)
# loc = положение легенды
```

Границы осей: set_xlim(), set_ylim()

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 200)
ax.plot(x, np.sin(x))
ax.set_xlim(-1, 11)  # (min, max)
ax.set_ylim(-1.5, 1.5)
# (min, max)
```

Все параметры сразу: ax.set(...)

```python
fig,ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10),
       ylim=(-2, 2),
       xlabel="x",
       ylabel="sin(x)",
       title="A simple Plot");
```

Размер картинки: figsize=...¶

```python
plot_small, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, np.sin(x));
```

Шрифты: rcParams

```python
mpl.rcParams.update({"font.size":12,"font.family":"serif"})# параметры для всех надписей
# mpl.rcParams['lines.linewidth'] = 2
mpl.rc("lines",linewidth=2,color="c")
x=np.linspace(1,10,50)
y=x**2
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title");
```
